```python
Brand=[]
Product_desc=[]
price=[]
discount=[]
```

```python
#Extract the tag having the brand-titles

for page in range(0,4):
    brands_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")

for i in brands_tags:
    brands=i.text
    Brand.append(brands)
driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
Brand
```

```python
len(Brand)
```

```python
#Extract the tag having the product_desc-titles

for page in range(0,4):
    products_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")

    
for i in products_tags:
    products=i.text
    Product_desc.append(products)
driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
Product_desc
```

```python
#Extract the tag having the price titles

for page in range(0,4):
    prices_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")

    
for i in prices_tags:
    prices=i.text
    price.append(prices)
driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
price
```

```python
#Extract the tag having the discount-titles

for page in range(0,4):
    discounts_tags=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")

    
for i in discounts_tags:
    discounts=i.text
    discount.append(discounts)
driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
discount
```
