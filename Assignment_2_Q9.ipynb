```python
Brand=[]
Shoe_description=[]
Price=[]
```

```python
#Extract the tag having the brand-titles
brands_tags=driver.find_elements_by_xpath("//h3[@class='product-brand']")
brands_tags[0:10]
```

```python
#Now need to extract the text from these tags

for page in range(0,3):
    brands_tags=driver.find_elements_by_xpath("//h3[@class='product-brand']")


for i in brands_tags:
    brands=i.text
    Brand.append(brands)
driver.find_element_by_xpath("//span[@class='pagination-arrowRight']").click()
Brand[0:100]
```

```python
driver.find_element_by_xpath("//li[@class='pagination-number']").click()
```

```python
#Extract the tag having the description-titles

description_tags=driver.find_elements_by_xpath("//h4[@class='product-product']")
description_tags[0:10]
```

```python
#Now need to extract the text from these tags

for page in range(0,3):
    description_tags=driver.find_elements_by_xpath("//h4[@class='product-product']")


for i in description_tags:
    description=i.text
    Shoe_description.append(description)
driver.find_element_by_xpath("//span[@class='pagination-arrowRight']").click()
Shoe_description[0:100]
```

```python
driver.find_element_by_xpath("//li[@class='pagination-number']").click()
```

```python
#Extract the tag having the Price-titles

price_tags=driver.find_elements_by_xpath("//span[@class='product-discountedPrice']")
price_tags[0:10]
```

```python
for page in range(0,2):
    price_tags=driver.find_elements_by_xpath("//span[@class='product-discountedPrice']")


for i in price_tags:
    price=i.text
    Price.append(price)
driver.find_element_by_xpath("//span[@class='pagination-arrowRight']").click()
Price[0:100]
```
